```python
yellow_trips.sort_values(by=['pickup_datetime_date'], inplace = True)
weather.sort_values(by=['date'], inplace = True)
```

```python
print(yellow_trips.shape)
yellow_trips_weather = pd.merge_asof(yellow_trips, weather, direction='nearest',
                                          left_on = 'pickup_datetime_date',
                                          right_on = 'date')
print(yellow_trips_weather.shape)
```

```python
yellow_trips_weather['pickup_month_dt'] = pd.to_datetime(yellow_trips_weather['pickup_datetime']).dt.to_period('M')
yellow_trips_weather['pickup_date'] = pd.to_datetime(yellow_trips_weather['pickup_datetime']).dt.to_period('d')
yellow_trips_weather['pickup_hour'] = pd.to_datetime(yellow_trips_weather['pickup_datetime']).dt.hour
yellow_trips_weather['pickup_weekday'] = yellow_trips_weather['pickup_datetime'].dt.dayofweek
yellow_trips_weather['weekend'] = np.where(yellow_trips_weather['pickup_weekday'] <= 4, False, True)
yellow_trips_weather['rush'] = np.where(((yellow_trips_weather['pickup_hour'] >= 6) & (yellow_trips_weather['pickup_hour'] <= 9)) |
                               ((yellow_trips_weather['pickup_hour'] >= 15) & (yellow_trips_weather['pickup_hour'] <= 20)), True, False)
```

```python
yellow_trips_day = yellow_trips_weather.groupby(['pickup_date'])[['max_temp', 'min_temp', 'avg_temp', 'precipitation', 'snowfall', 'snow_depth']].mean().reset_index()
yellow_trips_day['trips'] = yellow_trips_weather.groupby(['pickup_date'])['trip_distance'].mean().reset_index(name = 'trips')['trips']
yellow_trips_day.head()
```

```python
fig, ax = plt.subplots(figsize = plotsize)
m = sns.scatterplot(x="avg_temp", y="trips", data=yellow_trips_day, ax=ax)
ax.set(xlabel='Month', ylabel='Number of trips')
#ax.set_xticks(np.arange(0, 24,1))
plt.xticks(rotation=90);
plt.xticks
#plt.xlim(0, 0.1)
plt.title('Yellow Cabs - Trips per weekends by month');
path = 'Results'
plt.savefig(path + '/Images/delete.png')
```
