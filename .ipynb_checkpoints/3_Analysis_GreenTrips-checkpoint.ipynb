### Hipothesys testing

**Objective:** Compare the distance traveled between 201404 and 201406 with the one between 201504 and 201506

```python
path = 'Dataset/Parquets'
green_trips = pd.read_parquet(path + '/green_trips_filtered.parquet', engine = 'pyarrow')
green_trips['pickup_month_dt'] = pd.to_datetime(green_trips['pickup_datetime']).dt.to_period('M')
green_trips['pickup_date'] = pd.to_datetime(green_trips['pickup_datetime']).dt.to_period('d')
green_trips['pickup_hour'] = pd.to_datetime(green_trips['pickup_datetime']).dt.hour
green_trips['pickup_weekday'] = green_trips['pickup_datetime'].dt.dayofweek
green_trips['weekend'] = np.where(green_trips['pickup_weekday'] <= 4, False, True)
green_trips['rush'] = np.where(((green_trips['pickup_hour'] >= 6) & (green_trips['pickup_hour'] <= 9)) |
                               ((green_trips['pickup_hour'] >= 15) & (green_trips['pickup_hour'] <= 20)), True, False)
```

```python
green_trips.dtypes
```

```python
green_trips_201404_201406 = green_trips.loc[(green_trips['pickup_datetime'] >= '01/04/2014') & (green_trips['pickup_datetime'] <= '30/06/2014')]
green_trips_201504_201506 = green_trips.loc[(green_trips['pickup_datetime'] >= '01/04/2015') & (green_trips['pickup_datetime'] <= '30/06/2015')]
```

```python
green_trips_201404_201406.pickup_datetime.min()
green_trips_201404_201406.pickup_datetime.max()
```

**Hypothesis:** The distance of the trips that took place between 2014-04 and 2014-06 is statistically similar with the ones between 2015-04 and 2015-06

$$H_0: \mu_1=\mu_2$$
$$H_a: \mu_1 \neq \mu_2$$

```python
from scipy import stats
statistic, pvalue = stats.ttest_ind(green_trips_201404_201406.trip_distance, 
                                    green_trips_201504_201506.trip_distance, 
                                    equal_var=False)
statistic, pvalue
```

**Conclusion:** There is enough statistical evidence to assert that the distance of the trips that took place between 2014-04 and 2014-06 is statistically **different** with the ones between 2015-04 and 2015-06

**Hypothesis:** The distance of the trips that took place in weekends is statistically similar with the ones in weekdays

$$H_0: \mu_1=\mu_2$$
$$H_a: \mu_1 \neq \mu_2$$

```python
green_trips_weekend = green_trips.loc[(green_trips['weekend'] == True)]
green_trips_week = green_trips.loc[(green_trips['weekend'] == False)]

print(green_trips_weekend.trip_distance.mean())
print(green_trips_week.trip_distance.mean())
```

```python
from scipy import stats
statistic, pvalue = stats.ttest_ind(green_trips_weekend.trip_distance, 
                                    green_trips_week.trip_distance, 
                                    equal_var=False)
statistic, pvalue
```

**Conclusion:** There is enough statistical evidence to assert that the distance of the trips that took place in weekends is statistically **different** with the ones in weekdays

**Hypothesis:** The distance of the trips that took place in rush hours is statistically similar with the ones in non-rush hours

$$H_0: \mu_1=\mu_2$$
$$H_a: \mu_1 \neq \mu_2$$

```python
green_trips_rush = green_trips.loc[(green_trips['rush'] == True)]
green_trips_norush = green_trips.loc[(green_trips['rush'] == False)]
```

```python
from scipy import stats
statistic, pvalue = stats.ttest_ind(green_trips_rush.trip_distance, 
                                    green_trips_norush.trip_distance, 
                                    equal_var=False)
statistic, pvalue
```

**Conclusion:** There is enough statistical evidence to assert that the distance of the trips that took place in rush hours is statistically **different** with the ones in non-rush hours

# Maps for NTA analysis

```python
path = 'Dataset/Parquets'
green_trips = pd.read_parquet(path + '/green_trips_nta.parquet', engine = 'pyarrow')

# Opens the map as a GeoDataFrame
path = 'Dataset/Original'
nyc_bouroughs = geopandas.read_file(path + '/Neighborhood Tabulation Areas.geojson')
```

```python
green_trips_nta = green_trips.groupby('ntacode_pickup')['pickup_datetime'].count().reset_index(name = 'trips')
green_trips_nta.rename({'ntacode_pickup':'ntacode'}, inplace = True, axis = 1)

nyc_bouroughs_trips = nyc_bouroughs.merge(green_trips_nta, how = 'left', on = 'ntacode')
nyc_bouroughs_trips['trips'].fillna(0, inplace = True)
nyc_bouroughs_trips.head()
```

```python
# interactive visualization for beat-specific crime rate in 2017

m_trips = folium.Map(location=[40.738, -73.98],
                        zoom_start=13,
                        tiles="OpenStreetMap")

colormap = branca.colormap.LinearColormap(
    colors=['white','yellow','orange','red','darkred'],
    #index=beat_cn['count'].quantile([0.2,0.4,0.6,0.8]),b
    vmin=nyc_bouroughs_trips.trips.min(),
    vmax=nyc_bouroughs_trips.trips.max()
)

style_function = lambda x: {
    'fillColor': colormap(x['properties']['trips']),
    'color': 'black',
    'weight':2,
    'fillOpacity':0.5
}

stategeo = folium.GeoJson(
    nyc_bouroughs_trips.to_json(),
    name='Chicago beats',
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(
        fields=['ntaname', 'trips'],
        aliases=['NTA', 'Total trips'], 
        localize=True
    )
).add_to(m_trips)

colormap.add_to(m_trips)

path = 'Results'
m_trips.save(path + "/Maps/trips_nta_green_map.html")
```

```python
path = 'DataSet/Parquets'
nyc_bouroughs_trips.drop('geometry', axis = 1, inplace = True)
nyc_bouroughs_trips.to_parquet(path + '/nyc_bouroughs_trips_green.parquet', engine = 'pyarrow')
```
